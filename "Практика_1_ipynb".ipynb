1. **Гистограмма для числовых переменных:**

```python
sns.histplot(df['Sleep Duration'], kde=True)
plt.title('Распределение Sleep Duration')
plt.xlabel('Длительность сна')
plt.ylabel('Частота')
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
n_cols = len(num_cols)
n_rows = (n_cols + 2) // 3

fig, axes = plt.subplots(n_rows, 3, figsize=(18, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(df[col], kde=True, ax=axes[i])
    axes[i].set_title(f'Распределение {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

`Визуально нормальное распределение можно распознать по графику, который имеет форму колоколообразной, плавной симметричной кривой. Эта кривая имеет пик в центре, соответствующий среднему значению, и по мере удаления от центра высота кривой уменьшается, отражая уменьшение вероятности экстремальных значений.`

2. **Столбчатая диаграмма для категориальной переменной `Class`:**

```python
plt.figure(figsize=(14, 6))
sns.countplot(x='Occupation', data=df, order=df['Occupation'].value_counts().index)
plt.title('Распределение Occupation', fontsize=14)
plt.xlabel('Occupation', fontsize=12)
plt.ylabel('Количество', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns
n_cols = len(cat_cols)
n_rows = (n_cols + 2) // 3

fig, axes = plt.subplots(n_rows, 3, figsize=(18, 4 * n_rows))
axes = axes.flatten()

for ax, col in zip(axes, cat_cols):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')

    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты:**

```python
# Корреляционная матрица
numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_df.corr()

# Визуализация тепловой карты
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```
